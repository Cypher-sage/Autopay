```python
import datetime
date_time = datetime.datetime.now()
data['Age']=date_time.year - data['Year']
data.head()
```

```python
data.drop('Year',axis=1,inplace=True)
data.head()
```

```python
numerical_features = ['Present_Price', 'Kms_Driven', 'Owner', 'Age']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
sns.boxplot(data['Selling_Price'])
```

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
data.shape
```

```python
data.head(1)
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
data['Fuel_Type'].unique()
```

```python
data['Seller_Type'].unique()
```

```python
data['Seller_Type'] = data['Seller_Type'].map({'Dealer':0,'Individual':1})
data['Seller_Type'].unique()
```

```python
data['Transmission'].unique()
```

```python
data['Transmission'] =data['Transmission'].map({'Manual':0,'Automatic':1})
data['Transmission'].unique()
```

```python
data.head()
```

```python
X = data.drop(['Car_Name','Selling_Price'],axis=1)
y = data['Selling_Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
residuals1 = y_test - y_pred1
plt.figure(figsize=(8, 6))
sns.histplot(residuals1, kde=True, bins=30, color='orange')
plt.title('Residual Analysis - Linear Regression')
plt.xlabel('Residuals')
plt.show()
```

```python
mae1 = metrics.mean_absolute_error(y_test, y_pred1)
mse1 = metrics.mean_squared_error(y_test, y_pred1)
rmse1 = np.sqrt(mse1)
print(f'Linear Regression Model Metrics - MAE: {mae1}, MSE: {mse1}, RMSE: {rmse1}')
```

```python
residuals2 = y_test - y_pred2
plt.figure(figsize=(8, 6))
sns.histplot(residuals2, kde=True, bins=30, color='green')
plt.title('Residual Analysis - Random Forest')
plt.xlabel('Residuals')
plt.show()
```

```python
mae2 = metrics.mean_absolute_error(y_test, y_pred2)
mse2 = metrics.mean_squared_error(y_test, y_pred2)
rmse2 = np.sqrt(mse2)
print(f'Random Forest Model Metrics - MAE: {mae2}, MSE: {mse2}, RMSE: {rmse2}')
```

```python
residuals3 = y_test - y_pred3
plt.figure(figsize=(8, 6))
sns.histplot(residuals3, kde=True, bins=30, color='blue')
plt.title('Residual Analysis - Gradient Boosting')
plt.xlabel('Residuals')
plt.show()
```

```python
mae3 = metrics.mean_absolute_error(y_test, y_pred3)
mse3 = metrics.mean_squared_error(y_test, y_pred3)
rmse3 = np.sqrt(mse3)
print(f'Gradient Boosting Model Metrics - MAE: {mae3}, MSE: {mse3}, RMSE: {rmse3}')
```

```python
residuals4 = y_test - y_pred4
plt.figure(figsize=(8, 6))
sns.histplot(residuals4, kde=True, bins=30, color='purple')
plt.title('Residual Analysis - XGBoost')
plt.xlabel('Residuals')
plt.show()
```

```python
mae4 = metrics.mean_absolute_error(y_test, y_pred4)
mse4 = metrics.mean_squared_error(y_test, y_pred4)
rmse4 = np.sqrt(mse4)
print(f'XGBoost Model Metrics - MAE: {mae4}, MSE: {mse4}, RMSE: {rmse4}')
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
             "R2_SCORE":[score1,score2,score3,score4]})
```

```python
from tkinter import *
import joblib
import pandas as pd

def predict_car_price():
    try:
        # Get input values
        present_price = float(present_price_entry.get())
        kms_driven = float(kms_driven_entry.get())
        fuel_type = fuel_type_var.get()
        seller_type = seller_type_var.get()
        transmission = transmission_var.get()
        owner = owner_var.get()
        age = float(age_entry.get())

        # Load the trained model
        model = joblib.load('car_price_predictor')

        # Create a DataFrame with input values
        data_new = pd.DataFrame({
            'Present_Price': present_price,
            'Kms_Driven': kms_driven,
            'Fuel_Type': fuel_type,
            'Seller_Type': seller_type,
            'Transmission': transmission,
            'Owner': owner,
            'Age': age
        }, index=[0])

        # Predict car price
        result = model.predict(data_new)

        # Display result
        result_label.config(text=f"Car Purchase Amount: {result[0]:,.2f} Lakhs", font=("Helvetica", 16, "bold"), fg="green")

    except ValueError:
        result_label.config(text="Invalid input. Please enter numeric values.", font=("Helvetica", 12), fg="red")

master = Tk()
master.title("Car Price Prediction App")

# Frame for input fields
input_frame = Frame(master, padx=20, pady=20)
input_frame.grid(row=1, column=0)

# Labels for input fields
labels = ["Present Price (Lakhs)", "Kms Driven", "Fuel Type", "Seller Type", "Transmission", "Owner", "Age (Years)"]
for i, label_text in enumerate(labels):
    Label(input_frame, text=label_text, font=("Helvetica", 14)).grid(row=i, column=0, pady=5, sticky="e")

# Input fields
present_price_entry = Entry(input_frame, font=("Helvetica", 14))
kms_driven_entry = Entry(input_frame, font=("Helvetica", 14))
fuel_type_var = StringVar(master)
fuel_type_var.set("Petrol")  # Default value
OptionMenu(input_frame, fuel_type_var, "Petrol", "Diesel", "CNG").grid(row=2, column=1, pady=5, sticky="w")
seller_type_var = StringVar(master)
seller_type_var.set("Dealer")  # Default value
OptionMenu(input_frame, seller_type_var, "Dealer", "Individual").grid(row=3, column=1, pady=5, sticky="w")
transmission_var = StringVar(master)
transmission_var.set("Manual")  # Default value
OptionMenu(input_frame, transmission_var, "Manual", "Automatic").grid(row=4, column=1, pady=5, sticky="w")
owner_var = StringVar(master)
owner_var.set("First Owner")  # Default value
OptionMenu(input_frame, owner_var, "First Owner", "Second Owner", "Third Owner or More").grid(row=5, column=1, pady=5, sticky="w")
age_entry = Entry(input_frame, font=("Helvetica", 14))

# Place input fields
input_entries = [present_price_entry, kms_driven_entry, age_entry]
for i, entry in enumerate(input_entries):
    entry.grid(row=i, column=1, pady=5, padx=10)

# Button to predict car price
predict_button = Button(master, text='Predict', command=predict_car_price, font=("Helvetica", 16, "bold"), bg="green", fg="white")
predict_button.grid(row=2, column=0, pady=20)

# Label to display result
result_label = Label(master, text="", font=("Helvetica", 18, "italic"))
result_label.grid(row=3, column=0, pady=20)

mainloop()
```
